```python
#Importing modules
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#Integration using trapezoidal methods
def trapezeVect (f,a,b,n):
	x = np.linspace(a,b,n+1)
	y = f(x)
	h = (b-a)/n
	s = np.sum(y[:-1]+y[1:])*h/2
	return s
```

```python
#definition of distances
def Dp(z,f,a,b,n,c=3e5) : #Comoving distance
	return c * trapezeVect(f,a,b,n)
def Da(z,f,a,b,n,c=3e5) : #Angular distance
	return c/(1+z) * trapezeVect(f,a,b,n)
def Dl(z,f,a,b,n,c=3e5) : #luminosity distance
	return c*(1+z)*trapezeVect(f,a,b,n)
```
